```python
x = [i for i in range(1, 11)]
y = [10, 20, 30, 40, 50, 50, 40, 30, 20, 10]
plt.figure(figsize = (12, 5))
plt.plot(x,y, marker = "o",color = "green", mfc = "red")
plt.title("This is a line chart", fontsize = 25, color = "red")
plt.xlabel("X label", fontsize = 25, color = "green")
plt.ylabel("Y Label", fontsize = 25, color = "blue")
plt.show()
```
